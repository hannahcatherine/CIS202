# Iris Classification

In this activity, we will construct a machine learning model, specifically a decision tree classifier, to correctly classify iris flowers into different species based on the measurements of their sepals and petals. This classification task typically involves identifying three different species: Setosa, Versicolor, and Virginica.

```python
# Loading Libraries
import numpy as np  # Importing NumPy for numerical operations and array manipulations.
import pandas as pd  # Importing Pandas for data manipulation and analysis.

import matplotlib.pyplot as plt  # Importing Matplotlib for plotting and visualizations.

from sklearn.model_selection import train_test_split  # Importing the function to split data into training and testing sets.
from sklearn.tree import DecisionTreeClassifier  # Importing the Decision Tree Classifier for classification tasks.
from sklearn.metrics import accuracy_score  # Importing the accuracy score function to measure the accuracy of the model.
from sklearn.tree import plot_tree # Visualize the decision tree
```

# 1. Load the Iris data

```python
# Loading the iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

df = pd.read_csv(url, names=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)', 'Species'])

print(df) # To view the data
```

# 2. Data Preprocessing

When preparing to build a machine learning model, we split our dataset into two parts: training data, used to teach the model, and testing data, used to evaluate how well it performs. Here, X holds all our input data except the 'Species' column, while y contains the 'Species' labels. Using train_test_split, we divide our data into X_train and X_test for features, and y_train and y_test for labels, with 30% set aside for testing.

```python
# Splitting the data into train and test sets
X = df.drop("Species",axis=1) # Features of iris
y = df["Species"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# 3. Decision Tree Classification

Next, we create a DecisionTreeClassifier object called dtc and train our model using X_train (our training data) and y_train (training labels). Once trained, we use the model to predict y_pred for the test data (X_test). This helps us assess how well our model can predict the iris species based on its training.

```python
# Defining an object for DTC and train the model
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)

# Test the model
y_pred = dtc.predict(X_test)
```

```python
print('Accuracy of Decision Tree-Test: ', accuracy_score(y_pred, y_test))
```

This code creates a visual representation of our decision tree. First, it sets the size of the picture to make sure it's big enough to see clearly. Then, it draws the tree using the plot_tree function, showing how the tree makes decisions based on different features (like petal length and sepal width). Each node is colored to show which species of iris is most common at that point. Finally, it displays the tree.

```python
plt.figure(figsize=(12, 8))  # Adjust the figure size if needed
plot_tree(dtc, feature_names=X.columns, class_names=dtc.classes_, filled=True, impurity=False)
plt.show()
```
